```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib.request as req
url="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
res=req.urlopen(url)
soup=BeautifulSoup(res,'html.parser')
title=soup.find("title").string
wf=soup.find("wf").string
print(title)
print(wf)
```

```python
locations=soup.find_all("location")
for location in locations:
    city=location.find('city')
    print(city.text)
    datas=location.find_all('data')
    for data in datas:
#        print(data[0].text, data[1].text, data[2].text, data[3].text)
        tmEf=data.find('tmef').text
        wf=data.find('wf').text
        tmn=data.find('tmn').text
        tmx=data.find('tmx').text
        print(tmEf, wf, tmn, tmx)
```

```python
locations=soup.select("location")
#print(locations)
for loc in locations:
    city=loc.select_one('city').text
    print(city)
    datas=loc.select('data')
    for data in datas:
        #print(data.text)
        tmef=data.select('tmef')[0].text
        wf=data.select('wf')[0].text
        tmn=data.select('tmn')[0].text
        tmx=data.select('tmx')[0].text
        print(tmef, wf, tmn, tmx)
```

```python
def readPDF(pdfFile):
    rsmgr=PDFResourceManager()
    retstr=StringIO()
    laparams=LAParams()
    
    device=TextConverter(rsmgr, retstr, laparams=laparams)
    process_pdf(rsmgr, device, pdfFile)
    content=retstr.getvalue()
    retstr.close()
    return content
```

```python
pdfFile=urlopen("http://www.pythonscraping.com/pages/warandpeace/chapter1.pdf")
outputString=readPDF(pdfFile)
#print(outputString)
with open("data/pdf/result.txt",'w') as f:
    f.write(outputString)
    print("저장성공")
```
