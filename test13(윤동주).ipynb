```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
def func(url):
    res=req.urlopen(url)
    soup=BeautifulSoup(res,'html.parser')
    contents=soup.select('div.mw-parser-output p')
    if len(contents)>0:
        for con in contents:
            print(con.text)
```

```python
base="https://ko.wikisource.org"
url="https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res=req.urlopen(url)
soup=BeautifulSoup(res,'html.parser')
li_list=soup.select("div.mw-parser-output >ul > li")
for li in li_list:
    print(' 제목: ',li.a.string)
    s_li=li.find_all('li')
    if len(s_li)>0:
        for s in s_li:
            print('- 제목: ',s.a.string)
            #print(base+s.a['href'])
            func(base+s.a['href'])   
    else:
       #print(base+li.a['href'])
        func(base+li.a['href'])
```
