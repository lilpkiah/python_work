```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
def func(url):
    res=req.urlopen(url)
    soup=BeautifulSoup(res,'html.parser')
    content=soup.select('#mw-content-text > div.mw-parser-output p')
    if len(content)>0:
        print(content[0].text)
```

```python
def func(url):
    res=req.urlopen(url)
    soup=BeautifulSoup(res,'html.parser')
    content=soup.select('#mw-content-text > div.mw-parser-output p')
    if con in content:
        print(con.text)
```

```python
base="https://ko.wikisource.org"
url="https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res=req.urlopen(url)
soup=BeautifulSoup(res,'html.parser')
li_list=soup.select("div.mw-parser-output > ul > li")
for li in li_list:
    #제목 출력
    print('제목:', li.a.string)
    s_li=li.find_all('li')
    if len(s_li)>0:
        for s in s_li:
            print('- 제목: ',s.a.string)
            #print(base+s.a['href'])
            # 내용 출력
            res=req.urlopen(base+s.a['href'])
            soup=BeautifulSoup(res,'html.parser')
            content=soup.select('#mw-content-text > div.mw-parser-output p')
            if len(content)>0:
                print(content[0].text)
    else:
        #print(base+li.a['href'])
        res=req.urlopen(base+li.a['href'])
        soup=BeautifulSoup(res,'html.parser')
        content=soup.select('#mw-content-text > div.mw-parser-output p')[0].text
        if len(content)>0:
            print(content[0].text)
```
