```python
html=urlopen("http://www.yes24.com/Product/Search?domain=ALL&query=python")
soup=bs(html,'html.parser')
books=soup.select('#yesSchList > li')
#print(books)
i=1
for book in books:
    title=book.select('div > div.item_info > div.info_row.info_name > a.gd_name')[0].string
    title_list.append(title)
    image=book.select('div.item_img > div.img_canvas > span > span > a > em > img')[0]
    img_src=image['data-original']
    #save_image(img_src, i)
    
    #obj=book.select('div > div.item_info > div.info_row.info_pubGrp')
    #print(obj[0].text)
    #obj_arr=obj[0].text.split('\n')
    #print(obj_arr)
    
    author=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a')[0].string
    author_list.append(author)
    pub=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_pub > a')[0].string
    pub_list.append(pub)
    ym=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_date')
    if len(ym)>0:
        ymstr=ym[0].string
    else:
        ymstr="-"
    ym_list.append(ymstr)
    price=book.select('div.item_info > div.info_row.info_price > strong > em')[0].string
    price_list.append(price)
    book_list.append(tuple([title,author,pub,ymstr,price]))
    i+=1
    
print(book_list)
```
